```python
class RailwayReservationSystem:
    def __init__(self):
        self.booked_tickets = []
        self.load_data()
        self.root = tk.Tk()
        self.root.title("Railway Reservation System")
        self.setup_ui()

    def setup_ui(self):
        tk.Label(self.root, text="Name:").pack()
        self.name_entry = tk.Entry(self.root)
        self.name_entry.pack()

        tk.Label(self.root, text="Train:").pack()
        self.train_entry = tk.Entry(self.root)
        self.train_entry.pack()

        book_button = tk.Button(self.root, text="Book Ticket", command=self.book_ticket)
        book_button.pack()

        view_button = tk.Button(self.root, text="View Booked Tickets", command=self.view_tickets)
        view_button.pack()

        cancel_button = tk.Button(self.root, text="Cancel Ticket", command=self.cancel_ticket)
        cancel_button.pack()

    def book_ticket(self):
        name = self.name_entry.get()
        train = self.train_entry.get()
        if name and train:
            self.booked_tickets.append({'name': name, 'train': train})
            self.save_data()
            messagebox.showinfo("Success", "Ticket booked successfully!")
            self.name_entry.delete(0, tk.END)
            self.train_entry.delete(0, tk.END)
        else:
            messagebox.showerror("Error", "Please enter both name and train")

    def view_tickets(self):
        display_text = "\n".join([f"Name: {ticket['name']}, Train: {ticket['train']}" for ticket in self.booked_tickets])
        messagebox.showinfo("Booked Tickets", display_text if display_text else "No tickets booked")

    def cancel_ticket(self):
        name = simpledialog.askstring("Cancel Ticket", "Enter your name:")
        train = simpledialog.askstring("Cancel Ticket", "Enter train name:")
        ticket = {'name': name, 'train': train}
        if ticket in self.booked_tickets:
            self.booked_tickets.remove(ticket)
            self.save_data()
            messagebox.showinfo("Success", "Ticket cancelled successfully!")
        else:
            messagebox.showerror("Error", "Ticket not found")

    def save_data(self):
        with open('booked_tickets.json', 'w') as file:
            json.dump(self.booked_tickets, file)

    def load_data(self):
        if os.path.exists('booked_tickets.json'):
            with open('booked_tickets.json', 'r') as file:
                self.booked_tickets = json.load(file)

    def run(self):
        self.root.mainloop()
```

```python
system = RailwayReservationSystem()
system.run()
```
